```python
#if we done one hot encoding, no of columns will be7
#One hot encoding
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
col_trans=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['State']),remainder='passthrough')
x=col_trans.fit_transform(x)
x
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

Actual value--Pred value---df

```python
df1=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred})
df1
```

```python
df2=pd.DataFrame(x)
df2
```

```python
print("intercept is",model.intercept_)
print("coefficient or slope is",model.coef_)
list(zip(df2,model.coef_))
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("error is",mean_absolute_error(y_test,y_pred))
print("MAPE",mean_absolute_percentage_error(y_test,y_pred))
print("mSE",mean_squared_error(y_test,y_pred))
print("r2 score",r2_score(y_test,y_pred))
```
